```python
# Create 3x1 subplots
fig, axs = plt.subplots(1, 2, figsize=(5.3, 2.5))

fontsize = 12

# # ---------------- Silicon ----------------
# material = 'silicon'
# data_dft = np.loadtxt(globals()[material + '_dft'])
# k = np.unique(data_dft[:, 0])
# kbands = np.reshape(data_dft[:, 1], (-1, len(k)))
# k = k / max(k)

# for band in range(len(kbands)):
#     axs[0].plot(k, kbands[band, :] - globals()[material + '_efermi'],
#                 '-', linewidth=5, alpha=0.5, color='red', markersize=2)

# # PAOFLOW bands
# eband = np.load(globals()[material + '_pao'], allow_pickle=False)
# for ib in range(eband.shape[1]):
#     axs[0].plot(np.linspace(0, 1, eband.shape[0]), eband[:, ib],
#                 '--', color='black', markersize=0.5)

# axs[0].plot([], [], '-r', label='DFT')
# axs[0].plot([], [], '--k', label='PAOFLOW')

# # High symmetry k-points
# axs[0].axvline(0, linewidth=0.75, color='k', alpha=0.5)
# axs[0].axvline(1, linewidth=0.75, color='k', alpha=0.5)

# axs[0].set_xlim(0, 1)
# axs[0].set_ylabel("Energy (eV)", fontsize=10)
# axs[0].set_ylim(-5, 5)
# axs[0].set_xticks([])  # remove x-axis ticks for top subplot
# axs[0].tick_params(axis='y',labelsize=10)

# ---------------- Zirconia ----------------
material = 'zirconia'
data_dft = np.loadtxt(globals()[material + '_dft'])
k = np.unique(data_dft[:, 0])
kbands = np.reshape(data_dft[:, 1], (-1, len(k)))
k = k / max(k)

for band in range(len(kbands)):
    axs[1].plot(k, kbands[band, :] - globals()[material + '_efermi'],
                '-', linewidth=5, alpha=0.5, color='red', markersize=2)

# PAOFLOW bands
eband = np.load(globals()[material + '_pao'], allow_pickle=False)
for ib in range(eband.shape[1]):
    axs[1].plot(np.linspace(0, 1, eband.shape[0]), eband[:, ib],
                '--', color='black', markersize=0.5)

axs[1].plot([], [], '-r',label='DFT')
axs[1].plot([], [], '--k',label='PAOFLOW')

axs[1].axvline(0, linewidth=0.75, color='k', alpha=0.5)
axs[1].axvline(1, linewidth=0.75, color='k', alpha=0.5)

axs[1].set_xlim(0, 1)
axs[1].set_ylabel("Energy (eV)", fontsize=10)
axs[1].set_ylim(-5, 5)
axs[1].set_xticks([0, 1])
axs[1].set_xticklabels([r'$\Gamma$', 'X'], fontsize=10)
axs[1].tick_params(labelsize=10)
axs[1].tick_params(axis='y',labelsize=10)
axs[1].legend(bbox_to_anchor=(1,0.8),fontsize=fontsize-3)
# ---------------- HfO2 ----------------
material = 'hfo2'
data_dft = np.loadtxt(globals()[material + '_dft'])
k = np.unique(data_dft[:, 0])
kbands = np.reshape(data_dft[:, 1], (-1, len(k)))
k = k / max(k)

for band in range(len(kbands)):
    axs[0].plot(k, kbands[band, :] - globals()[material + '_efermi'],
                '-', linewidth=5, alpha=0.5, color='red', markersize=2)

# PAOFLOW bands
eband = np.load(globals()[material + '_pao'], allow_pickle=False)
for ib in range(eband.shape[1]):
    axs[0].plot(np.linspace(0, 1, eband.shape[0]), eband[:, ib],
                '--', color='black', markersize=0.5)

axs[0].plot([], [], '-r',label='DFT')
axs[0].plot([], [], '--k',label='PAOFLOW')

axs[0].axvline(0, linewidth=0.75, color='k', alpha=0.5)
axs[0].axvline(1, linewidth=0.75, color='k', alpha=0.5)

axs[0].set_xlim(0, 1)
axs[0].set_ylabel("Energy (eV)", fontsize=10)
axs[0].set_ylim(-5, 5)
axs[0].tick_params(axis='y',labelsize=10)
# Only bottom subplot gets the X ticks
axs[0].set_xticks([0, 1])
axs[0].set_xticklabels([r'$\Gamma$', 'X'], fontsize=10)
axs[0].legend(bbox_to_anchor=(1,0.8),fontsize=fontsize-3)


# ---------------- Legend & Layout ----------------
# fig.legend(ncols=2, loc="lower center", bbox_to_anchor=(0.57, -0.032),fontsize=fontsize-2)
fig.tight_layout()
plt.savefig("bands.pdf", bbox_inches="tight")
```

```python
material = 'silicon'

data_dft = np.loadtxt(globals()[material + '_dft'])

k = np.unique(data_dft[:,0])
kbands = np.reshape(data_dft[:, 1], (-1,len(k)))
k = k/max(k)

for band in range(len(kbands)):
    plt.plot(k, kbands[band,:] - globals()[material + '_efermi'], '-',linewidth=5, alpha=0.5, color='red', markersize=2)
    # plt.plot(k, kbands[band,:], '-',linewidth=5, alpha=0.5, color='red', markersize=2)

# plot the paoflow bands
eband = np.load(globals()[material + '_pao'], allow_pickle=False)
for ib in range(eband.shape[1]):
    plt.plot(np.linspace(0,1,eband.shape[0]),eband[:,ib],'--',color='black', markersize=0.5)


plt.plot([],[],'-r', label='DFT')
plt.plot([],[],'--k', label='PAOFLOW')
plt.legend()

# High symmetry k-points (check bands_pp.out)
plt.axvline(0, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1, linewidth=0.75, color='k', alpha=0.5)


plt.xlim(0, 1)
plt.ylabel("Energy (eV)")
plt.ylim(-5,5)

# Annotate the band gap
plt.tight_layout()
plt.show()   
```
