```python
import time
import os
import numpy as np 
from pathlib import Path
import argparse
import ffsim
import ffsim
import pyscf

from sqd_lattice.pre_process import make_pyscf_slater_det,make_two_body_tensor
from sqd_lattice.util import load_pickle,load_yaml,save_pickle,load_json,save_json
```

```python
material = "platinum_2"
ne = "30"

folder_path = f"../runs/{material}/{ne}e"

# Loading data
base_config = load_yaml(f"../runs/{material}","base_config.yaml")
data = load_pickle(f"../runs/{material}","data_dict.pkl")
results_pre_process = load_pickle(folder_path,"results_pre_process.pkl")
```

```python
# Building Hamiltonian in ffsim
hamiltonian = ffsim.MolecularHamiltonian(one_body_tensor=data["hopping_matrix"],two_body_tensor=data["two_body_tensor"])
hamiltonian = hamiltonian.rotated(results_pre_process["mo_coeff"].T) # Changing to molecular
```

```python
linear_op_ffsim = ffsim.linear_operator(hamiltonian,norb=data["num_orbitals"],nelec=results_pre_process["ne_ab"])
slater_det_state = ffsim.hartree_fock_state(data["num_orbitals"],results_pre_process["ne_ab"])
slater_det_energy = np.real(np.vdot(slater_det_state, linear_op_ffsim @ slater_det_state))
slater_det_energy
```
